```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', 200)
```

```python
df_A_daily = pd.read_excel("data_folder/data.xlsx")
df_A_interval = pd.read_excel("data_folder/data.xlsx", sheet_name="A - Interval")
df_B_daily = pd.read_excel("data_folder/data.xlsx", sheet_name="B - Daily")
df_B_interval = pd.read_excel("data_folder/data.xlsx", sheet_name="B - Interval")
df_C_daily = pd.read_excel("data_folder/data.xlsx", sheet_name="C - Daily")
df_C_interval = pd.read_excel("data_folder/data.xlsx", sheet_name="C - Interval")
df_D_daily = pd.read_excel("data_folder/data.xlsx", sheet_name="D - Daily")
df_D_interval = pd.read_excel("data_folder/data.xlsx", sheet_name="D - Interval") 
```

```python

df_A_daily_removed = df_A_daily.dropna()
df_B_daily_removed = df_B_daily.dropna()
df_C_daily_removed = df_C_daily.dropna()
df_D_daily_removed = df_D_daily.dropna()
df_A_interval_removed = df_A_interval.dropna()
df_B_interval_removed = df_B_interval.dropna()
df_C_interval_removed = df_C_interval.dropna()
df_D_interval_removed = df_D_interval.dropna()
```

```python
cv_means_a = df_A_interval_removed.groupby("Interval")["Call Volume"].mean()
cct_means_a = df_A_interval_removed.groupby("Interval")["CCT"].mean()
abd_means_a = df_A_interval_removed.groupby("Interval")["Abandoned Rate"].mean()

cv_means_b = df_B_interval_removed.groupby("Interval")["Call Volume"].mean()
cct_means_b = df_B_interval_removed.groupby("Interval")["CCT"].mean()
abd_means_b = df_B_interval_removed.groupby("Interval")["Abandoned Rate"].mean()

cv_means_c = df_C_interval_removed.groupby("Interval")["Call Volume"].mean()
cct_means_c = df_C_interval_removed.groupby("Interval")["CCT"].mean()
abd_means_c = df_C_interval_removed.groupby("Interval")["Abandoned Rate"].mean()

cv_means_d = df_D_interval_removed.groupby("Interval")["Call Volume"].mean()
cct_means_d = df_D_interval_removed.groupby("Interval")["CCT"].mean()
abd_means_d = df_D_interval_removed.groupby("Interval")["Abandoned Rate"].mean()
```

```python
df_template = pd.read_csv("data_folder/template_forecast_v00.csv")

df_template_removed = df_template.copy()
```

```python
def convert_HHMM(t):
    return f"{t.hour}:{t.minute:02d}"
```

```python
cv_means_a.index = cv_means_a.index.map(convert_HHMM)
cct_means_a.index = cct_means_a.index.map(convert_HHMM)
abd_means_a.index = abd_means_a.index.map(convert_HHMM)

cv_means_b.index = cv_means_b.index.map(convert_HHMM)
cct_means_b.index = cct_means_b.index.map(convert_HHMM)
abd_means_b.index = abd_means_b.index.map(convert_HHMM)

cv_means_c.index = cv_means_c.index.map(convert_HHMM)
cct_means_c.index = cct_means_c.index.map(convert_HHMM)
abd_means_c.index = abd_means_c.index.map(convert_HHMM)

cv_means_d.index = cv_means_d.index.map(convert_HHMM)
cct_means_d.index = cct_means_d.index.map(convert_HHMM)
abd_means_d.index = abd_means_d.index.map(convert_HHMM)
```

```python
df_template_removed["Calls_Offered_A"] = df_template_removed["Interval"].map(cv_means_a).round(0).astype(int)
df_template_removed["CCT_A"] = df_template_removed["Interval"].map(cct_means_a).round(0).astype(int)
df_template_removed["Abandoned_Rate_A"] = df_template_removed["Interval"].map(abd_means_a)
df_template_removed["Abandoned_Calls_A"] = df_template_removed["Calls_Offered_A"] * df_template_removed["Abandoned_Rate_A"]
df_template_removed["Abandoned_Calls_A"] = (df_template_removed["Abandoned_Calls_A"].round(0).astype(int))
df_template_removed["Abandoned_Rate_A"] = (df_template_removed["Abandoned_Rate_A"] * 100).round(2).astype(str) + "%"

df_template_removed["Calls_Offered_B"] = df_template_removed["Interval"].map(cv_means_b).round(0).astype(int)
df_template_removed["CCT_B"] = df_template_removed["Interval"].map(cct_means_b).round(0).astype(int)
df_template_removed["Abandoned_Rate_B"] = df_template_removed["Interval"].map(abd_means_b)
df_template_removed["Abandoned_Calls_B"] = df_template_removed["Calls_Offered_B"] * df_template_removed["Abandoned_Rate_B"]
df_template_removed["Abandoned_Calls_B"] = (df_template_removed["Abandoned_Calls_B"].round(0).astype(int))
df_template_removed["Abandoned_Rate_B"] = (df_template_removed["Abandoned_Rate_B"] * 100).round(2).astype(str) + "%"

df_template_removed["Calls_Offered_C"] = df_template_removed["Interval"].map(cv_means_c).round(0).astype(int)
df_template_removed["CCT_C"] = df_template_removed["Interval"].map(cct_means_c).round(0).astype(int)
df_template_removed["Abandoned_Rate_C"] = df_template_removed["Interval"].map(abd_means_c)
df_template_removed["Abandoned_Calls_C"] = df_template_removed["Calls_Offered_C"] * df_template_removed["Abandoned_Rate_C"]
df_template_removed["Abandoned_Calls_C"] = (df_template_removed["Abandoned_Calls_C"].round(0).astype(int))
df_template_removed["Abandoned_Rate_C"] = (df_template_removed["Abandoned_Rate_C"] * 100).round(2).astype(str) + "%"

df_template_removed["Calls_Offered_D"] = df_template_removed["Interval"].map(cv_means_d).round(0).astype(int)
df_template_removed["CCT_D"] = df_template_removed["Interval"].map(cct_means_d).round(0).astype(int)
df_template_removed["Abandoned_Rate_D"] = df_template_removed["Interval"].map(abd_means_d)
df_template_removed["Abandoned_Calls_D"] = df_template_removed["Calls_Offered_D"] * df_template_removed["Abandoned_Rate_D"]
df_template_removed["Abandoned_Calls_D"] = (df_template_removed["Abandoned_Calls_D"].round(0).astype(int))
df_template_removed["Abandoned_Rate_D"] = (df_template_removed["Abandoned_Rate_D"] * 100).round(2).astype(str) + "%"

df_template_removed
```

```python
df_template_removed.to_csv("forecast_v01.csv", index = False)
```

```python
# from sklearn.linear_model import PoissonRegressor
# from sklearn.linear_model import LinearRegression
# from sklearn.linear_model import LogisticRegression

import statsmodels.formula.api as smf
```

```python
import datetime

def convert_datetime(s):
    hour, min = map(int, s.split(":")) 
    return datetime.time(hour, min)
```

```python
df_basic_regression = df_template.copy()
```

```python
# convert call_volume to int (initially float)
df_A_interval_removed['Call Volume'] = df_A_interval_removed['Call Volume'].astype(int)

# poisson regression based on the interval and day
model_a_cv = smf.poisson("Q('Call Volume') ~ C(Interval) + C(Day)", 
                                       data=df_A_interval_removed).fit()

# convert str to datetime 
df_basic_regression['Interval'] = df_basic_regression['Interval'].apply(convert_datetime)

# fill column with predicted values of template df
df_basic_regression["Calls_Offered_A"] = model_a_cv.predict(df_basic_regression).round(0).astype(int)
df_basic_regression
```

```python
# plt.figure(figsize=(18, 6))
# sns.scatterplot(x="Interval", y="CCT", data=df_A_interval_removed)
# plt.xticks(rotation=45)
# plt.show()
```

```python
df_A_interval_removed["Calls_Offered_A"] = (df_A_interval_removed["Call Volume"] - df_A_interval_removed["Abandoned Calls"]).round(0).astype(int)
df_A_interval_removed
```

```python
model_a_cct = smf.ols("Q('CCT') ~ C(Interval) + C(Day) + Q('Calls_Offered_A')", 
                      data=df_A_interval_removed).fit()

df_basic_regression["CCT_A"] = model_a_cct.predict(df_basic_regression)
df_basic_regression
```

```python
import statsmodels.api as sm
```

```python
model_a_abd = smf.glm("Q('Abandoned Rate') ~ C(Interval) + C(Day)",
                        data=df_A_interval_removed, family = sm.families.Binomial()).fit()

df_basic_regression["Abandoned_Rate_A"] = model_a_abd.predict(df_basic_regression)
df_basic_regression
```

```python
df_basic_regression["Abandoned_Calls_A"] = df_basic_regression["Calls_Offered_A"] * df_basic_regression["Abandoned_Rate_A"]
df_basic_regression["Abandoned_Calls_A"] = (df_basic_regression["Abandoned_Calls_A"].round(0).astype(int))

df_basic_regression
```
